```python
import pandas as pd
import psycopg2

# Replace these with your actual database connection details
db_params = {
    'dbname': 'week1',
    'user': 'postgres',
    'password': 'habte',
    'host': 'localhost',
    'port': '5432'
}

# Open a connection to the PostgreSQL database
conn = psycopg2.connect(**db_params)

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Example: execute a SELECT query
cursor.execute("SELECT * FROM xdr_data;")

# Fetch all rows
rows = cursor.fetchall()

# Get column names
column_names = [desc[0] for desc in cursor.description]

# Create a Pandas DataFrame
df = pd.DataFrame(rows, columns=column_names)

# Print the DataFrame
print(df)

# Don't forget to close the cursor and connection when done
cursor.close()
conn.close()
```

```python
print(df.head)
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()
```

```python
df_cleaned
```

```python
# Fill missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)
```

```python
df_filled 
```

```python
# Convert 'Start' column to datetime
df['Start'] = pd.to_datetime(df['Start'])
```

```python
df['Start']
```

```python
# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
```

```python
df_no_duplicates
```

```python
# Remove leading and trailing whitespaces
df['Last Location Name'] = df['Last Location Name'].str.strip()
```

```python
# Replace certain strings
df['Handset Manufacturer'] = df['Handset Manufacturer'].replace('Sm-', 'Samsung ', regex=True)
```

```python
df['Handset Manufacturer']
```

```python
# Rename columns
df = df.rename(columns={'Bearer Id': 'Bearer_ID', 'MSISDN/Number': 'MSISDN'})
```

```python
df
```

```python
# Identify outliers in a specific column (e.g., 'Dur. (ms)')
outliers = df['Dur. (ms)'].quantile(0.99)
df_no_outliers = df[df['Dur. (ms)'] < outliers]
```

```python
df_no_outliers 
```

```python
# Summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
# Univariate Analysis
import matplotlib.pyplot as plt

# Histogram for 'Dur. (ms)'
plt.hist(df['Dur. (ms)'], bins=30, edgecolor='black')
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.title('Distribution of Duration')
plt.show()
```

```python
#categorical variable
import seaborn as sns

# Bar plot for 'Handset Manufacturer'
plt.figure(figsize=(10, 6))
sns.countplot(x='Handset Manufacturer', data=df)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Handset Manufacturer')
plt.ylabel('Count')
plt.title('Distribution of Handset Manufacturer')
plt.show()
```

```python
#bivarite analysis
# Scatter plot for 'Dur. (ms)' vs 'Avg RTT DL (ms)'
plt.scatter(df['Dur. (ms)'], df['Avg RTT DL (ms)'])
plt.xlabel('Duration (ms)')
plt.ylabel('Avg RTT DL (ms)')
plt.title('Scatter Plot: Duration vs Avg RTT DL')
plt.show()
```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
correlation_matrix = numeric_columns.corr()

# Heatmap for correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box plot for 'Dur. (ms)'
plt.boxplot(df['Dur. (ms)'])
plt.xlabel('Duration (ms)')
plt.title('Box Plot: Duration')
plt.show()
```

```python
# Univariate analysis - Histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['Dur. (ms)'], bins=30, kde=True)
plt.title('Distribution of Duration')
plt.show()
```

```python
# Bivariate analysis - Scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Avg RTT DL (ms)', y='Avg RTT UL (ms)', data=df)
plt.title('Scatter Plot of Avg RTT DL vs Avg RTT UL')
plt.show()
```
